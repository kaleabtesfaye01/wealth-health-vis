```python
import pandas as pd

gdp = pd.read_csv('../data/raw/gdp-per-capita-worldbank.csv')
life_expectancy = pd.read_csv('../data/raw/life-expectancy.csv')
child_mortality = pd.read_csv('../data/raw/child-mortality.csv')
human_development_index = pd.read_csv('../data/raw/human-development-index.csv')

gdp.info()
```

```python
gdp.drop(columns=['World region according to OWID'], inplace=True)

gdp[gdp['Year'] >= 2020]['Year'].value_counts()
```

```python
life_expectancy.info()
```

```python
life_expectancy.dropna(inplace=True)

life_expectancy[life_expectancy['Year'] >= 2020]['Year'].value_counts()
```

```python
child_mortality.info()
```

```python
child_mortality.dropna(inplace=True)

child_mortality[child_mortality['Year'] >= 2020]['Year'].value_counts()
```

```python
human_development_index.info()
```

```python
human_development_index[human_development_index['Year'] >= 2020]['Year'].value_counts()
```

2022 Year Chosen


```python
gdp_2022 = gdp[gdp['Year'] == 2022].drop(columns=['Year'])
life_expectancy_2022 = life_expectancy[life_expectancy['Year'] == 2022].drop(columns=['Year'])
child_mortality_2022 = child_mortality[child_mortality['Year'] == 2022].drop(columns=['Year'])
human_development_index_2022 = human_development_index[human_development_index['Year'] == 2022].drop(columns=['Year', 'World region according to OWID'])

gdp_2022.info()
```

```python
life_expectancy_2022.info()
```

```python
child_mortality_2022.info()
```

```python
human_development_index_2022.info()
```

```python
preprocessed_df = gdp_2022.merge(life_expectancy_2022, on=['Code', 'Entity'], how='inner')
preprocessed_df = preprocessed_df.merge(child_mortality_2022, on=['Code', 'Entity'], how='inner')
preprocessed_df = preprocessed_df.merge(human_development_index_2022, on=['Code', 'Entity'], how='inner')

preprocessed_df.info()
```

```python
preprocessed_df.to_csv('../data/processed/level1_data.csv', index=False)
```
